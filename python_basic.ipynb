```python
import csv

file_path = "grades.csv"

students = [
    ["김철수", 85],
    ["이영희", 92],
    ["박민수", 78],
    ["최수진", 95]
]

with open(file_path, "w", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["name", "score"])
    writer.writerows(students)

print(f"학생 성적이 {file_path}에 저장되었습니다.\n")

total = 0
count = 0

print("성적 분석 결과:")

with open(file_path, "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        name = row["name"]
        score = int(row["score"])
        print(f"{name}: {score}점")
        total += score
        count += 1

average = total / count
print(f"전체 평균: {average}점")
```

## 8. 모듈과 패키지 (3문제)
### 문제 8-1

```python
import math

def circle_area(r):
    return math.pi * (r ** 2)

def rectangle_area(width, height):
    return width * height

def factorial(n):
    result = 1
    for i in range(n, 0, -1):
        result *= i
    return result

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(f"원의 넓이: {circle_area(5):.2f}")
print(f"직사각형 넓이: {rectangle_area(10, 5)}")
print(f"팩토리얼 5!: {factorial(5)}")
print(f"최대공약수(48, 18) = {gcd(48, 18)}")
```

### 문제 8-2

```python
import datetime
import random

now = datetime.datetime(2025, 7, 20, 14, 30, 25)
print(f"현재 날짜와 시간: {now}")

formatted = now.strftime("%Y년 %m월 %d일 %A")

day_map = {
    "Monday": "월요일", "Tuesday": "화요일", "Wednesday": "수요일",
    "Thursday": "목요일", "Friday": "금요일", "Saturday": "토요일", "Sunday": "일요일"
}
for eng, kor in day_map.items():
    formatted = formatted.replace(eng, kor)

print(f"포맷된 날짜: {formatted}")

random.seed(100)

rand_int = random.randint(1, 100)
print(f"임의의 숫자: {rand_int}")

rand_float = round(random.uniform(1, 100), 2)
print(f"임의의 실수: {rand_float}")

fruits = ["사과", "바나나", "오렌지", "포도", "딸기"]
rand_fruit = random.choice(fruits)
print(f"임의의 리스트 요소: {rand_fruit}")

random.shuffle(fruits)
print(f"섞인 리스트: {fruits}")
```
