# 1. 웹스크래핑 연습문제
## 1-1. Daum 뉴스기사 제목 스크래핑하기
### 질문1 :  아래의 url에서 뉴스기사의 링크와 제목을 출력하세요.
- 다음 경제 뉴스 URL
- url = 'https://news.daum.net/economy'


```python
# reqeusts, bs4 import
import requests
import bs4

# BeautifulSoup 클래스 import
from bs4 import BeautifulSoup
```

```python
# 스크래핑할 url 선언
url = 'https://news.daum.net/economy'

# dictionary 타입으로 요청 parameter 설정
req_param = {
    'category': 'economy'
}
url = 'https://news.daum.net/{category}'.format(**req_param)
print(url)

# 요청 헤더 설정 : 브라우저 정보 (크롤링 차단 방지)
req_header = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
}

# requests의 get() 함수 호출하기
res = requests.get(url, headers=req_header)
res.encoding = 'utf-8'
print(type(res))
print(res.status_code)
# print('---------------------------------')
# BeautifulSoup 객체 생성
if res.ok:
    soup = BeautifulSoup(res.text,'html.parser')
    li_tags = soup.select("ul.list_newsheadline2 li")
    print(type(li_tags),len(li_tags))
#    print(li_tags)
    for li_tag in li_tags:
        a_tag = li_tag.find('a')
        link = a_tag['href']
        print(link)
        strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
        title = strong_tag.text
        print(title)
# 응답(response)이 Error 이면 status code 출력    
else:
    print(f'Error Code = {res.status_code}')
```

### 질문2:  여러개의 section 중 하나를 선택해서 url에서 뉴스기사의 링크와 제목을 출력하는 코드를 함수로 작성하기

```python
# dictionary 타입으로 section 설정
section_dict = {
    "기후/환경": "climate",
    "사회": "society",
    "경제": "economy",
    "정치": "politics",
    "국제": "world",
    "문화": "culture",
    "생활": "life",
    "IT/과학": "tech",
    "인물": "people",
}

# 선택한 section의 뉴스기사의 링크와 제목 출력 함수인 print_news를 정의
def print_news(section_name):
    category = section_dict.get(section_name, "기후/환경")
    url = f"https://news.daum.net/{category}"
    print('======>'f"{url} {section_name} 뉴스 <======")
    req_header = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
    }
    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'
    if res.ok:
        soup = BeautifulSoup(res.text,'html.parser')
        li_tags = soup.select("ul.list_newsheadline2 li")
        print(type(li_tags),len(li_tags))
        for li_tag in li_tags:
            a_tag = li_tag.find('a')
            link = a_tag['href']
            print(link)
            strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
            title = strong_tag.text
            print(title)
    # 응답(response)이 Error 이면 status code 출력    
    else:
        print(f'Error Code = {res.status_code}')
```

```python
print_news('경제')
print_news('사회')
```
